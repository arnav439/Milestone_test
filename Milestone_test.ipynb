```python
## write a program that display appropriate message at the road crossing i.e., red/stop,yellow/wait,green/go
def display_traffic_message(signal_color):
    # Convert the input to lowercase to handle case-insensitivity
    signal_color = signal_color.lower()
    
    if signal_color == 'red':
        print("STOP: The signal is red. Please stop your vehicle.")
    elif signal_color == 'yellow':
        print("WAIT: The signal is yellow. Prepare to stop or proceed with caution.")
    elif signal_color == 'green':
        print("GO: The signal is green. You may proceed.")
    else:
        print("INVALID: The color you entered is not a valid traffic signal color.")

def main():
    # Prompt user for the color of the signal
    signal_color = input("Enter the color of the traffic signal (red, yellow, green): ").strip()
    
    # Display the appropriate message based on the color
    display_traffic_message(signal_color)

if __name__ == '__main__':
    main()
```

```python
#write a program to create a calculate which do only four calculations +-*/ in python

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y != 0:
        return x / y
    else:
        return "Error: Division by zero is not allowed."

def main():
    print("Simple Calculator")
    print("Select operation:")
    print("1. Add (+)")
    print("2. Subtract (-)")
    print("3. Multiply (*)")
    print("4. Divide (/)")

    # Take user input for operation
    choice = input("Enter the number corresponding to the operation: ")

    # Take user input for numbers
    try:
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))
    except ValueError:
        print("Invalid input! Please enter numeric values.")
        return

    # Perform the chosen operation
    if choice == '1':
        print(f"The result is: {add(num1, num2)}")
    elif choice == '2':
        print(f"The result is: {subtract(num1, num2)}")
    elif choice == '3':
        print(f"The result is: {multiply(num1, num2)}")
    elif choice == '4':
        result = divide(num1, num2)
        print(f"The result is: {result}")
    else:
        print("Invalid choice! Please select a valid operation.")

if __name__ == '__main__':
    main()
```

```python
#write  the program to find the larger of the three pre-specified number using ternary operations 
def find_largest(a, b, c):
    # Use nested ternary operations to determine the largest of the three numbers
    largest = a if (a > b and a > c) else (b if (b > c) else c)
    return largest

def main():
    # Pre-specified numbers
    num1 = 10
    num2 = 20
    num3 = 15
    
    # Find the largest number
    largest_number = find_largest(num1, num2, num3)
    
    # Print the result
    print(f"The largest of {num1}, {num2}, and {num3} is {largest_number}")

if __name__ == '__main__':
    main()
```

```python
#write a program to find the factors of whole number using while loop
def find_factors(number):
    # Initialize the divisor
    divisor = 1
    
    # List to store factors
    factors = []
    
    # Use a while loop to find factors
    while divisor <= number:
        if number % divisor == 0:
            factors.append(divisor)
        divisor += 1
    
    return factors

def main():
    # Take input from the user
    try:
        num = int(input("Enter a whole number: "))
        if num < 0:
            print("Please enter a non-negative whole number.")
            return
    except ValueError:
        print("Invalid input! Please enter a valid integer.")
        return
    
    # Find factors
    factors = find_factors(num)
    
    # Print the factors
    print(f"Factors of {num} are: {', '.join(map(str, factors))}")

if __name__ == '__main__':
    main()
```

```python
# write a program to find the prime number between 2 to 100 using nested for loops .

def is_prime(number):
    # Handle cases for numbers less than 2
    if number <= 1:
        return False
    # Check for factors from 2 to the square root of the number
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True

def find_primes(start, end):
    primes = []
    for num in range(start, end + 1):
        if is_prime(num):
            primes.append(num)
    return primes

def main():
    start = 2
    end = 1000
    primes = find_primes(start, end)
    
    print(f"Prime numbers between {start} and {end} are:")
    print(primes)

if __name__ == '__main__':
    main()
```

```python
"""1.12. Write the programs for the following:
. Accept the marks of the student in five major subjects and display the same.
· Calculate the sum of the marks of all subjects.Divide the total marks by number of subjects (i.e. 5), calculate
percentage = total marks/5 and display the percentage.
. Find the grade of the student as per the following criteria . Hint: Use Match & case for this .:
Criteria
Grade
percentage > 85
A
percentage < 85 && percentage >= 75
B
percentage < 75 && percentage >= 50
C
percentage > 30 && percentage <= 50
D
percentage <30
Reappear"""


def main():
    # Input marks for five subjects
    try:
        marks = []
        for i in range(1, 6):
            mark = float(input(f"Enter the marks for subject {i}: "))
            if mark < 0:
                print("Marks cannot be negative. Please enter valid marks.")
                return
            marks.append(mark)

        # Display the marks
        print("\nMarks in each subject are:")
        for i, mark in enumerate(marks, start=1):
            print(f"Subject {i}: {mark}")

        # Calculate total marks
        total_marks = sum(marks)

        # Calculate percentage
        number_of_subjects = len(marks)
        percentage = total_marks / number_of_subjects

        # Display percentage
        print(f"\nTotal Marks: {total_marks}")
        print(f"Percentage: {percentage:.2f}%")

        # Determine the grade using match-case
        match percentage:
            case p if p > 85:
                grade = "A"
            case p if 75 <= p <= 85:
                grade = "B"
            case p if 50 <= p < 75:
                grade = "C"
            case p if 30 <= p < 50:
                grade = "D"
            case _:
                grade = "Reappear"
        
        # Display the grade
        print(f"Grade: {grade}")

    except ValueError:
        print("Invalid input! Please enter numeric values for marks.")

if __name__ == '__main__':
    main()
```

```python
"""1.13. Write a program for VIBGYOR Spectrum based on their Wavelength using.
Wavelength Range:
COLOR
WAVELENGTH (nm)
Violet
400.0-440.0
Indigo
440.0-460.0
Blue
460.0-500.0
Green
500.0-570.0
Yellow
570.0-590.0
Orange
590.0-620.0
Red
620.0-720.0"""

def determine_color(wavelength):
    # Determine the color based on the wavelength
    match wavelength:
        case _ if 400.0 <= wavelength < 440.0:
            return "Violet"
        case _ if 440.0 <= wavelength < 460.0:
            return "Indigo"
        case _ if 460.0 <= wavelength < 500.0:
            return "Blue"
        case _ if 500.0 <= wavelength < 570.0:
            return "Green"
        case _ if 570.0 <= wavelength < 590.0:
            return "Yellow"
        case _ if 590.0 <= wavelength < 620.0:
            return "Orange"
        case _ if 620.0 <= wavelength <= 720.0:
            return "Red"
        case _:
            return "Out of range"

def main():
    try:
        # Input wavelength from the user
        wavelength = float(input("Enter the wavelength in nanometers (nm): "))

        # Check if the wavelength is within a plausible range
        if wavelength < 0:
            print("Wavelength cannot be negative. Please enter a valid wavelength.")
            return

        # Determine the color
        color = determine_color(wavelength)
        
        # Display the result
        print(f"The color corresponding to the wavelength {wavelength} nm is {color}.")
    
    except ValueError:
        print("Invalid input! Please enter a numeric value for the wavelength.")

if __name__ == '__main__':
    main()
```

```python
"""1.14.Consider the gravitational interactions between the Earth, Moon, and Sun in our solar system.
Given:
mass_earth = 5.972e24 # Mass of Earth in kilograms
mass_moon = 7.34767309e22 # Mass of Moon in kilograms
mass_sun = 1.989e30 # Mass of Sun in kilograms
distance_earth_sun = 1.496ell # Average distance between Earth and Sun in meters
distance_moon_earth = 3.844e8 # Average distance between Moon and Earth in meters
Tasks:
. Calculate the gravitational force between the Earth and the Sun.
. Calculate the gravitational force between the Moon and the Earth.
· Compare the calculated forces to determine which gravitational force is stronger.
. Explain which celestial body (Earth or Moon) is more attracted to the other based on the comparison."""




def calculate_gravitational_force(m1, m2, distance):
    G = 6.67430e-11  # Gravitational constant in m^3 kg^-1 s^-2
    return G * m1 * m2 / (distance ** 2)

def main():
    # Given data
    mass_earth = 5.972e24  # kg
    mass_moon = 7.34767309e22  # kg
    mass_sun = 1.989e30  # kg
    distance_earth_sun = 1.496e11  # meters
    distance_moon_earth = 3.844e8  # meters

    # Calculate gravitational force between Earth and Sun
    force_earth_sun = calculate_gravitational_force(mass_earth, mass_sun, distance_earth_sun)
    
    # Calculate gravitational force between Moon and Earth
    force_moon_earth = calculate_gravitational_force(mass_moon, mass_earth, distance_moon_earth)
    
    # Print the forces
    print(f"Gravitational force between Earth and Sun: {force_earth_sun:.2e} N")
    print(f"Gravitational force between Moon and Earth: {force_moon_earth:.2e} N")
    
    # Compare the forces
    if force_earth_sun > force_moon_earth:
        print("The gravitational force between Earth and the Sun is stronger.")
        print("This means the Earth is more strongly attracted to the Sun than the Moon is to the Earth.")
    elif force_earth_sun < force_moon_earth:
        print("The gravitational force between the Moon and the Earth is stronger.")
        print("This means the Moon is more strongly attracted to the Earth than the Earth is to the Sun.")
    else:
        print("The gravitational forces between the two pairs are equal.")

if __name__ == '__main__':
    main()
```

```python
"""2. Design and implement a Python program for managing student information using object-oriented
principles. Create a class called `Student` with encapsulated attributes for name, age, and roll number.
Implement getter and setter methods for these attributes. Additionally, provide methods to display student
information and update student details.
Tasks:
· Define the `Student` class with encapsulated attributes.
· Implement getter and setter methods for the attributes.
· Write methods to display student information and update details.
· Create instances of the `Student` class and test the implemented functionality."""

class Student:
    def __init__(self, name, age, roll_number):
        self.__name = name           # Encapsulated attribute for name
        self.__age = age             # Encapsulated attribute for age
        self.__roll_number = roll_number # Encapsulated attribute for roll number
    
    # Getter and Setter for name
    def get_name(self):
        return self.__name
    
    def set_name(self, name):
        self.__name = name
    
    # Getter and Setter for age
    def get_age(self):
        return self.__age
    
    def set_age(self, age):
        if age > 0:  # Ensure age is positive
            self.__age = age
        else:
            print("Age must be positive.")
    
    # Getter and Setter for roll number
    def get_roll_number(self):
        return self.__roll_number
    
    def set_roll_number(self, roll_number):
        self.__roll_number = roll_number
    
    # Method to display student information
    def display_info(self):
        print(f"Name: {self.__name}")
        print(f"Age: {self.__age}")
        print(f"Roll Number: {self.__roll_number}")
    
    # Method to update student details
    def update_details(self, name=None, age=None, roll_number=None):
        if name is not None:
            self.set_name(name)
        if age is not None:
            self.set_age(age)
        if roll_number is not None:
            self.set_roll_number(roll_number)

def main():
    # Create instances of the Student class
    student1 = Student("Alice", 20, "S12345")
    student2 = Student("Bob", 22, "S67890")
    
    # Display initial information
    print("Initial Student Information:")
    student1.display_info()
    student2.display_info()
    
    # Update student details
    print("\nUpdating student details...")
    student1.update_details(name="Alicia", age=21)
    student2.update_details(roll_number="S54321")
    
    # Display updated information
    print("\nUpdated Student Information:")
    student1.display_info()
    student2.display_info()

if __name__ == '__main__':
    main()

```

```python
"""3.Develop a Python program for managing library resources efficiently. Design a class named "LibraryBook
with attributes like book name, author, and availability status. Implement methods for borrowing and
returning books while ensuring proper encapsulation of attributes.
Tasks:
S
· 1. Create the `LibraryBook` class with encapsulated attributes.
· 2. Implement methods for borrowing and returning books.
· 3. Ensure proper encapsulation to protect book details.
· 4. Test the borrowing and returning functionality with sample data."""
class LibraryBook:
    def __init__(self, book_name, author):
        self.__book_name = book_name       # Encapsulated attribute for book name
        self.__author = author             # Encapsulated attribute for author
        self.__is_available = True         # Encapsulated attribute for availability status
    
    # Getter for book name
    def get_book_name(self):
        return self.__book_name
    
    # Getter for author
    def get_author(self):
        return self.__author
    
    # Getter for availability status
    def is_available(self):
        return self.__is_available
    
    # Method to borrow the book
    def borrow_book(self):
        if self.__is_available:
            self.__is_available = False
            print(f"You have successfully borrowed '{self.__book_name}' by {self.__author}.")
        else:
            print(f"'{self.__book_name}' by {self.__author} is currently not available.")
    
    # Method to return the book
    def return_book(self):
        if not self.__is_available:
            self.__is_available = True
            print(f"'{self.__book_name}' by {self.__author} has been successfully returned.")
        else:
            print(f"'{self.__book_name}' by {self.__author} was not borrowed.")
    
    # Method to display book details
    def display_book_info(self):
        status = "Available" if self.__is_available else "Not Available"
        print(f"Book Name: {self.__book_name}")
        print(f"Author: {self.__author}")
        print(f"Status: {status}")

def main():
    # Create instances of LibraryBook
    book1 = LibraryBook("To Kill a Mockingbird", "Harper Lee")
    book2 = LibraryBook("1984", "George Orwell")
    
    # Display initial information
    print("Initial Book Information:")
    book1.display_book_info()
    book2.display_book_info()
    
    # Borrow books
    print("\nAttempting to borrow books...")
    book1.borrow_book()
    book2.borrow_book()
    
    # Display updated information
    print("\nUpdated Book Information:")
    book1.display_book_info()
    book2.display_book_info()
    
    # Return books
    print("\nReturning books...")
    book1.return_book()
    book2.return_book()
    
    # Display final information
    print("\nFinal Book Information:")
    book1.display_book_info()
    book2.display_book_info()

if __name__ == '__main__':
    main()

```

```python
"""4.Create a simple banking system using object-oriented concepts in Python. Design classes representing
different types of bank accounts such as savings and checking. Implement methods for deposit, withdraw,
and balance inquiry. Utilize inheritance to manage different account types efficiently.
Tasks:
· 1. Define base class(es) for bank accounts with common attributes and methods.
· 2. Implement subclasses for specific account types (e.g ., SavingsAccount, CheckingAccount).
· 3. Provide methods for deposit, withdraw, and balance inquiry in each subclass.
. 4. Test the banking system by creating instances of different account types and performing transactions.
"""
# Base class for bank accounts
class BankAccount:
    def __init__(self, account_number, account_holder, balance=0.0):
        self.__account_number = account_number
        self.__account_holder = account_holder
        self.__balance = balance
    
    # Getter for balance
    def get_balance(self):
        return self.__balance
    
    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount:.2f}. New balance: ${self.__balance:.2f}.")
        else:
            print("Deposit amount must be positive.")
    
    # Method to withdraw money
    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.__balance:
                self.__balance -= amount
                print(f"Withdrew ${amount:.2f}. New balance: ${self.__balance:.2f}.")
            else:
                print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")
    
    # Method to display account information
    def display_account_info(self):
        print(f"Account Number: {self.__account_number}")
        print(f"Account Holder: {self.__account_holder}")
        print(f"Balance: ${self.__balance:.2f}")
        
# Subclass for Savings Account
class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder, balance=0.0, interest_rate=0.01):
        super().__init__(account_number, account_holder, balance)
        self.__interest_rate = interest_rate
    
    # Method to add interest to the balance
    def add_interest(self):
        interest = self.get_balance() * self.__interest_rate
        self.deposit(interest)
        print(f"Added interest: ${interest:.2f}.")
        
# Subclass for Checking Account
class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder, balance=0.0, overdraft_limit=1000.0):
        super().__init__(account_number, account_holder, balance)
        self.__overdraft_limit = overdraft_limit
    
    # Method to withdraw money with overdraft protection
    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.get_balance() + self.__overdraft_limit:
                new_balance = self.get_balance() - amount
                self._BankAccount__balance = new_balance  # Access protected attribute
                print(f"Withdrew ${amount:.2f}. New balance: ${self.get_balance():.2f}.")
            else:
                print("Withdrawal amount exceeds overdraft limit.")
        else:
            print("Withdrawal amount must be positive.")

def main():
    # Create instances of SavingsAccount and CheckingAccount
    savings = SavingsAccount("S123456", "Alice Johnson", 5000.0, 0.03)
    checking = CheckingAccount("C987654", "Bob Smith", 1000.0, 500.0)
    
    # Display initial information
    print("Initial Savings Account Information:")
    savings.display_account_info()
    
    print("\nInitial Checking Account Information:")
    checking.display_account_info()
    
    # Perform transactions on Savings Account
    print("\nPerforming transactions on Savings Account...")
    savings.deposit(200.0)
    savings.withdraw(150.0)
    savings.add_interest()
    
    # Perform transactions on Checking Account
    print("\nPerforming transactions on Checking Account...")
    checking.deposit(500.0)
    checking.withdraw(1200.0)
    checking.withdraw(300.0)  # Should be allowed with overdraft
    
    # Display updated information
    print("\nUpdated Savings Account Information:")
    savings.display_account_info()
    
    print("\nUpdated Checking Account Information:")
    checking.display_account_info()

if __name__ == '__main__':
    main()

```

```python
"""6. Write a code for Restaurant Management System Using OOPS:
· Create a Menultem class that has attributes such as name, description, price, and category.
· Implement methods to add a new menu item, update menu item information, and remove a menu item
from the menu.
· Use encapsulation to hide the menu item's unique identification number.
· Inherit from the Menultem class to create a FoodItem class and a Beverageltem class, each with their own
specific attributes and methods.
"""

import uuid

# Base class MenuItem
class MenuItem:
    def __init__(self, name, description, price, category):
        self.__id = uuid.uuid4()  # Unique ID for each menu item
        self.__name = name
        self.__description = description
        self.__price = price
        self.__category = category
    
    # Getters and Setters
    def get_id(self):
        return self.__id

    def get_name(self):
        return self.__name
    
    def set_name(self, name):
        self.__name = name
    
    def get_description(self):
        return self.__description
    
    def set_description(self, description):
        self.__description = description
    
    def get_price(self):
        return self.__price
    
    def set_price(self, price):
        if price >= 0:
            self.__price = price
        else:
            print("Price must be non-negative.")
    
    def get_category(self):
        return self.__category
    
    def set_category(self, category):
        self.__category = category
    
    # Method to display item details
    def display_info(self):
        print(f"ID: {self.__id}")
        print(f"Name: {self.__name}")
        print(f"Description: {self.__description}")
        print(f"Price: ${self.__price:.2f}")
        print(f"Category: {self.__category}")

# Subclass FoodItem
class FoodItem(MenuItem):
    def __init__(self, name, description, price, category, cuisine_type):
        super().__init__(name, description, price, category)
        self.__cuisine_type = cuisine_type
    
    # Getter and Setter for cuisine type
    def get_cuisine_type(self):
        return self.__cuisine_type
    
    def set_cuisine_type(self, cuisine_type):
        self.__cuisine_type = cuisine_type
    
    # Overriding display_info method to include cuisine type
    def display_info(self):
        super().display_info()
        print(f"Cuisine Type: {self.__cuisine_type}")

# Subclass BeverageItem
class BeverageItem(MenuItem):
    def __init__(self, name, description, price, category, volume_ml):
        super().__init__(name, description, price, category)
        self.__volume_ml = volume_ml
    
    # Getter and Setter for volume
    def get_volume_ml(self):
        return self.__volume_ml
    
    def set_volume_ml(self, volume_ml):
        if volume_ml > 0:
            self.__volume_ml = volume_ml
        else:
            print("Volume must be positive.")
    
    # Overriding display_info method to include volume
    def display_info(self):
        super().display_info()
        print(f"Volume: {self.__volume_ml} ml")

# Example usage
def main():
    # Creating instances of FoodItem and BeverageItem
    pizza = FoodItem("Margherita Pizza", "Classic pizza with cheese and tomato", 12.99, "Main Course", "Italian")
    cola = BeverageItem("Cola", "Carbonated soft drink", 1.99, "Beverage", 330)
    
    # Displaying item information
    print("Menu Item Information:")
    pizza.display_info()
    print()
    cola.display_info()
    
    # Updating item information
    print("\nUpdating Menu Item...")
    pizza.set_price(13.49)
    pizza.set_cuisine_type("Italian-American")
    cola.set_volume_ml(500)
    
    # Displaying updated item information
    print("\nUpdated Menu Item Information:")
    pizza.display_info()
    print()
    cola.display_info()

if __name__ == '__main__':
    main()
```

```python
"""7.Write a code for Hotel Management System using OOPS :
. Create a Room class that has attributes such as room number, room type, rate, and availability (private).
· Implement methods to book a room, check in a guest, and check out a guest.
· Use encapsulation to hide the room's unique identification number.
· Inherit from the Room class to create a SuiteRoom class and a StandardRoom class, each with their own
specific attributes and methods.
"""
class Room:
    def __init__(self, room_number, room_type, rate):
        self.__room_number = room_number    # Unique room number (encapsulated)
        self.__room_type = room_type        # Room type (e.g., Suite, Standard)
        self.__rate = rate                  # Room rate per night
        self.__availability = True          # Room availability status
        self.__guest_name = None            # Name of the guest (if checked in)
    
    # Getter for room number (public method)
    def get_room_number(self):
        return self.__room_number
    
    # Getter for room type
    def get_room_type(self):
        return self.__room_type
    
    def set_room_type(self, room_type):
        self.__room_type = room_type
    
    # Getter and setter for rate
    def get_rate(self):
        return self.__rate
    
    def set_rate(self, rate):
        if rate >= 0:
            self.__rate = rate
        else:
            print("Rate must be non-negative.")
    
    # Method to check availability
    def is_available(self):
        return self.__availability
    
    # Method to book a room
    def book_room(self, guest_name):
        if self.__availability:
            self.__availability = False
            self.__guest_name = guest_name
            print(f"Room {self.__room_number} has been booked for {guest_name}.")
        else:
            print(f"Room {self.__room_number} is not available.")
    
    # Method to check in a guest
    def check_in(self, guest_name):
        if self.__availability:
            print(f"Room {self.__room_number} is available. Booking for {guest_name}.")
            self.book_room(guest_name)
        else:
            if self.__guest_name == guest_name:
                print(f"Guest {guest_name} is already checked in.")
            else:
                print(f"Room {self.__room_number} is occupied.")
    
    # Method to check out a guest
    def check_out(self):
        if not self.__availability:
            print(f"Checking out guest {self.__guest_name} from room {self.__room_number}.")
            self.__availability = True
            self.__guest_name = None
        else:
            print(f"Room {self.__room_number} is already vacant.")
    
    # Method to display room information
    def display_info(self):
        status = "Available" if self.__availability else "Occupied"
        print(f"Room Number: {self.__room_number}")
        print(f"Room Type: {self.__room_type}")
        print(f"Rate: ${self.__rate:.2f} per night")
        print(f"Availability: {status}")
        if not self.__availability:
            print(f"Current Guest: {self.__guest_name}")

# Subclass SuiteRoom
class SuiteRoom(Room):
    def __init__(self, room_number, rate, has_jacuzzi):
        super().__init__(room_number, "Suite", rate)
        self.__has_jacuzzi = has_jacuzzi
    
    # Getter and Setter for jacuzzi feature
    def has_jacuzzi(self):
        return self.__has_jacuzzi
    
    def set_jacuzzi(self, has_jacuzzi):
        self.__has_jacuzzi = has_jacuzzi
    
    # Overriding display_info to include jacuzzi feature
    def display_info(self):
        super().display_info()
        print(f"Jacuzzi: {'Yes' if self.__has_jacuzzi else 'No'}")

# Subclass StandardRoom
class StandardRoom(Room):
    def __init__(self, room_number, rate, bed_count):
        super().__init__(room_number, "Standard", rate)
        self.__bed_count = bed_count
    
    # Getter and Setter for bed count
    def get_bed_count(self):
        return self.__bed_count
    
    def set_bed_count(self, bed_count):
        if bed_count > 0:
            self.__bed_count = bed_count
        else:
            print("Bed count must be positive.")
    
    # Overriding display_info to include bed count
    def display_info(self):
        super().display_info()
        print(f"Bed Count: {self.__bed_count}")

def main():
    # Create instances of SuiteRoom and StandardRoom
    suite = SuiteRoom("101", 250.0, True)
    standard = StandardRoom("202", 100.0, 2)
    
    # Display room information
    print("Initial Room Information:")
    suite.display_info()
    print()
    standard.display_info()
    
    # Book rooms
    print("\nBooking Rooms...")
    suite.check_in("John Doe")
    standard.check_in("Jane Smith")
    
    # Attempt to check in a guest in an already occupied room
    print("\nAttempting to check in to an already occupied room...")
    suite.check_in("Alice Johnson")
    
    # Display updated room information
    print("\nUpdated Room Information:")
    suite.display_info()
    print()
    standard.display_info()
    
    # Check out guests
    print("\nChecking out guests...")
    suite.check_out()
    standard.check_out()
    
    # Display final room information
    print("\nFinal Room Information:")
    suite.display_info()
    print()
    standard.display_info()

if __name__ == '__main__':
    main()
```

```python
"""8.Write a code for Fitness Club Management System using OOPS:
· Create a Member class that has attributes such as name, age, membership type, and membership status
(private).
· Implement methods to register a new member, renew a membership, and cancel a membership.
· Use encapsulation to hide the member's unique identification number.
· Inherit from the Member class to create a FamilyMember class and an IndividualMember class, each with
their own specific attributes and methods."""


import uuid

# Base class Member
class Member:
    def __init__(self, name, age, membership_type):
        self.__id = uuid.uuid4()  # Unique ID for each member (encapsulated)
        self.__name = name
        self.__age = age
        self.__membership_type = membership_type
        self.__membership_status = "Active"  # Default membership status
    
    # Getter for member ID (public method)
    def get_id(self):
        return self.__id
    
    # Getter and Setter for name
    def get_name(self):
        return self.__name
    
    def set_name(self, name):
        self.__name = name
    
    # Getter and Setter for age
    def get_age(self):
        return self.__age
    
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age must be positive.")
    
    # Getter and Setter for membership type
    def get_membership_type(self):
        return self.__membership_type
    
    def set_membership_type(self, membership_type):
        self.__membership_type = membership_type
    
    # Method to register a new member
    def register_member(self):
        if self.__membership_status == "Active":
            print(f"Member {self.__name} is already registered.")
        else:
            self.__membership_status = "Active"
            print(f"Member {self.__name} has been registered.")
    
    # Method to renew membership
    def renew_membership(self):
        if self.__membership_status == "Inactive":
            self.__membership_status = "Active"
            print(f"Membership for {self.__name} has been renewed.")
        else:
            print(f"Membership for {self.__name} is already active.")
    
    # Method to cancel membership
    def cancel_membership(self):
        if self.__membership_status == "Active":
            self.__membership_status = "Inactive"
            print(f"Membership for {self.__name} has been canceled.")
        else:
            print(f"Membership for {self.__name} is already inactive.")
    
    # Method to display member information
    def display_info(self):
        print(f"ID: {self.__id}")
        print(f"Name: {self.__name}")
        print(f"Age: {self.__age}")
        print(f"Membership Type: {self.__membership_type}")
        print(f"Membership Status: {self.__membership_status}")

# Subclass FamilyMember
class FamilyMember(Member):
    def __init__(self, name, age, membership_type, number_of_family_members):
        super().__init__(name, age, membership_type)
        self.__number_of_family_members = number_of_family_members
    
    # Getter and Setter for number of family members
    def get_number_of_family_members(self):
        return self.__number_of_family_members
    
    def set_number_of_family_members(self, number_of_family_members):
        if number_of_family_members > 0:
            self.__number_of_family_members = number_of_family_members
        else:
            print("Number of family members must be positive.")
    
    # Overriding display_info method to include number of family members
    def display_info(self):
        super().display_info()
        print(f"Number of Family Members: {self.__number_of_family_members}")

# Subclass IndividualMember
class IndividualMember(Member):
    def __init__(self, name, age, membership_type, personal_trainer):
        super().__init__(name, age, membership_type)
        self.__personal_trainer = personal_trainer
    
    # Getter and Setter for personal trainer
    def has_personal_trainer(self):
        return self.__personal_trainer
    
    def set_personal_trainer(self, personal_trainer):
        self.__personal_trainer = personal_trainer
    
    # Overriding display_info method to include personal trainer information
    def display_info(self):
        super().display_info()
        print(f"Personal Trainer: {'Yes' if self.__personal_trainer else 'No'}")

def main():
    # Create instances of FamilyMember and IndividualMember
    family_member = FamilyMember("Alice Johnson", 35, "Family", 4)
    individual_member = IndividualMember("Bob Smith", 28, "Individual", True)
    
    # Display member information
    print("Initial Member Information:")
    family_member.display_info()
    print()
    individual_member.display_info()
    
    # Register members
    print("\nRegistering Members...")
    family_member.register_member()
    individual_member.register_member()
    
    # Renew memberships
    print("\nRenewing Memberships...")
    family_member.renew_membership()
    individual_member.renew_membership()
    
    # Cancel memberships
    print("\nCanceling Memberships...")
    family_member.cancel_membership()
    individual_member.cancel_membership()
    
    # Display updated member information
    print("\nUpdated Member Information:")
    family_member.display_info()
    print()
    individual_member.display_info()

if __name__ == '__main__':
    main()
```

```python
"""9.Write a code for Event Management System using OOPS:
· Create an Event class that has attributes such as name, date, time, location, and list of attendees (private).
· Implement methods to create a new event, add or remove attendees, and get the total number of
attendees.
· Use encapsulation to hide the event's unique identification number.
. Inherit from the Event class to create a PrivateEvent class and a PublicEvent class, each with their own
specific attributes and methods.
"""
import uuid

# Base class Event
class Event:
    def __init__(self, name, date, time, location):
        self.__id = uuid.uuid4()  # Unique ID for each event (encapsulated)
        self.__name = name
        self.__date = date
        self.__time = time
        self.__location = location
        self.__attendees = []  # Private list to store attendees
    
    # Getter for event ID (public method)
    def get_id(self):
        return self.__id
    
    # Getter and Setter for event name
    def get_name(self):
        return self.__name
    
    def set_name(self, name):
        self.__name = name
    
    # Getter and Setter for event date
    def get_date(self):
        return self.__date
    
    def set_date(self, date):
        self.__date = date
    
    # Getter and Setter for event time
    def get_time(self):
        return self.__time
    
    def set_time(self, time):
        self.__time = time
    
    # Getter and Setter for event location
    def get_location(self):
        return self.__location
    
    def set_location(self, location):
        self.__location = location
    
    # Method to add an attendee
    def add_attendee(self, attendee_name):
        if attendee_name not in self.__attendees:
            self.__attendees.append(attendee_name)
            print(f"Attendee {attendee_name} has been added.")
        else:
            print(f"Attendee {attendee_name} is already on the list.")
    
    # Method to remove an attendee
    def remove_attendee(self, attendee_name):
        if attendee_name in self.__attendees:
            self.__attendees.remove(attendee_name)
            print(f"Attendee {attendee_name} has been removed.")
        else:
            print(f"Attendee {attendee_name} is not on the list.")
    
    # Method to get the total number of attendees
    def get_total_attendees(self):
        return len(self.__attendees)
    
    # Method to display event information
    def display_info(self):
        print(f"Event ID: {self.__id}")
        print(f"Name: {self.__name}")
        print(f"Date: {self.__date}")
        print(f"Time: {self.__time}")
        print(f"Location: {self.__location}")
        print(f"Total Attendees: {self.get_total_attendees()}")

# Subclass PrivateEvent
class PrivateEvent(Event):
    def __init__(self, name, date, time, location, access_code):
        super().__init__(name, date, time, location)
        self.__access_code = access_code
    
    # Getter and Setter for access code
    def get_access_code(self):
        return self.__access_code
    
    def set_access_code(self, access_code):
        self.__access_code = access_code
    
    # Overriding display_info method to include access code
    def display_info(self):
        super().display_info()
        print(f"Access Code: {self.__access_code}")

# Subclass PublicEvent
class PublicEvent(Event):
    def __init__(self, name, date, time, location, is_ticket_required):
        super().__init__(name, date, time, location)
        self.__is_ticket_required = is_ticket_required
    
    # Getter and Setter for ticket requirement
    def is_ticket_required(self):
        return self.__is_ticket_required
    
    def set_ticket_required(self, is_ticket_required):
        self.__is_ticket_required = is_ticket_required
    
    # Overriding display_info method to include ticket requirement
    def display_info(self):
        super().display_info()
        print(f"Ticket Required: {'Yes' if self.__is_ticket_required else 'No'}")

def main():
    # Create instances of PrivateEvent and PublicEvent
    private_event = PrivateEvent("Private Dinner", "2024-08-15", "19:00", "Penthouse Suite", "PRV1234")
    public_event = PublicEvent("Public Concert", "2024-09-10", "21:00", "Central Park", True)
    
    # Display event information
    print("Initial Event Information:")
    private_event.display_info()
    print()
    public_event.display_info()
    
    # Add and remove attendees
    print("\nManaging Attendees...")
    private_event.add_attendee("Alice Johnson")
    private_event.add_attendee("Bob Smith")
    private_event.remove_attendee("Alice Johnson")
    
    public_event.add_attendee("Carol Davis")
    public_event.add_attendee("David Evans")
    
    # Display updated event information
    print("\nUpdated Event Information:")
    private_event.display_info()
    print()
    public_event.display_info()

if __name__ == '__main__':
    main()
```

```python
"""10.Write a code for Airline Reservation System using OOPS:
· Create a Flight class that has attributes such as flight number, departure and arrival airports, departure and
arrival times, and available seats (private).
· Implement methods to book a seat, cancel a reservation, and get the remaining available seats.
· Use encapsulation to hide the flight's unique identification number.
· Inherit from the Flight class to create a DomesticFlight class and an InternationalFlight class, each with their
own specific attributes and methods.
"""
import uuid
from datetime import datetime

# Base class Flight
class Flight:
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats):
        self.__id = uuid.uuid4()  # Unique ID for each flight (encapsulated)
        self.__flight_number = flight_number
        self.__departure_airport = departure_airport
        self.__arrival_airport = arrival_airport
        self.__departure_time = departure_time
        self.__arrival_time = arrival_time
        self.__total_seats = total_seats
        self.__available_seats = total_seats  # Initially, all seats are available
    
    # Getter for flight ID (public method)
    def get_id(self):
        return self.__id
    
    # Getter and Setter for flight number
    def get_flight_number(self):
        return self.__flight_number
    
    def set_flight_number(self, flight_number):
        self.__flight_number = flight_number
    
    # Getter and Setter for departure airport
    def get_departure_airport(self):
        return self.__departure_airport
    
    def set_departure_airport(self, departure_airport):
        self.__departure_airport = departure_airport
    
    # Getter and Setter for arrival airport
    def get_arrival_airport(self):
        return self.__arrival_airport
    
    def set_arrival_airport(self, arrival_airport):
        self.__arrival_airport = arrival_airport
    
    # Getter and Setter for departure time
    def get_departure_time(self):
        return self.__departure_time
    
    def set_departure_time(self, departure_time):
        self.__departure_time = departure_time
    
    # Getter and Setter for arrival time
    def get_arrival_time(self):
        return self.__arrival_time
    
    def set_arrival_time(self, arrival_time):
        self.__arrival_time = arrival_time
    
    # Method to book a seat
    def book_seat(self):
        if self.__available_seats > 0:
            self.__available_seats -= 1
            print(f"Seat booked successfully. Remaining seats: {self.__available_seats}")
        else:
            print("No available seats to book.")
    
    # Method to cancel a reservation
    def cancel_reservation(self):
        if self.__available_seats < self.__total_seats:
            self.__available_seats += 1
            print(f"Reservation canceled successfully. Remaining seats: {self.__available_seats}")
        else:
            print("No reservations to cancel.")
    
    # Method to get the remaining available seats
    def get_remaining_seats(self):
        return self.__available_seats
    
    # Method to display flight information
    def display_info(self):
        print(f"Flight ID: {self.__id}")
        print(f"Flight Number: {self.__flight_number}")
        print(f"Departure Airport: {self.__departure_airport}")
        print(f"Arrival Airport: {self.__arrival_airport}")
        print(f"Departure Time: {self.__departure_time}")
        print(f"Arrival Time: {self.__arrival_time}")
        print(f"Total Seats: {self.__total_seats}")
        print(f"Available Seats: {self.__available_seats}")

# Subclass DomesticFlight
class DomesticFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats, state):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats)
        self.__state = state
    
    # Getter and Setter for state
    def get_state(self):
        return self.__state
    
    def set_state(self, state):
        self.__state = state
    
    # Overriding display_info method to include state
    def display_info(self):
        super().display_info()
        print(f"State: {self.__state}")

# Subclass InternationalFlight
class InternationalFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats, country):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats)
        self.__country = country
    
    # Getter and Setter for country
    def get_country(self):
        return self.__country
    
    def set_country(self, country):
        self.__country = country
    
    # Overriding display_info method to include country
    def display_info(self):
        super().display_info()
        print(f"Country: {self.__country}")

def main():
    # Create instances of DomesticFlight and InternationalFlight
    domestic_flight = DomesticFlight("DL123", "JFK", "LAX", "2024-08-15 07:00", "2024-08-15 10:00", 150, "California")
    international_flight = InternationalFlight("BA456", "LHR", "JFK", "2024-09-10 16:00", "2024-09-10 19:00", 300, "United States")
    
    # Display flight information
    print("Domestic Flight Information:")
    domestic_flight.display_info()
    print()
    print("International Flight Information:")
    international_flight.display_info()
    
    # Book and cancel seats
    print("\nBooking Seats...")
    domestic_flight.book_seat()
    domestic_flight.book_seat()
    
    print("\nCanceling Reservation...")
    domestic_flight.cancel_reservation()
    
    print("\nBooking Seats for International Flight...")
    international_flight.book_seat()
    
    # Display updated flight information
    print("\nUpdated Domestic Flight Information:")
    domestic_flight.display_info()
    print()
    print("Updated International Flight Information:")
    international_flight.display_info()

if __name__ == '__main__':
    main()
```
